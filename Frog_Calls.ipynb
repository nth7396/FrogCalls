# Data Transformation

```python
from sklearn.preprocessing import StandardScaler

# Perform data normalization
scaler = StandardScaler()
df_normalized = scaler.fit_transform(df[mfcc_columns])

# Create a new DataFrame with the normalized values and 'Family' column
df_normalized = pd.DataFrame(df_normalized, columns=mfcc_columns)
df_final = pd.concat([df_normalized, df['Family']], axis=1)

# Print the final DataFrame
df_final.head(10)
```

# K-Means

```python
from sklearn.cluster import KMeans
import warnings

# Extract the MFCC columns from the df_final DataFrame
X = df_final[mfcc_columns]

# Define the range of cluster numbers
k_values = range(1, 11)

# Initialize a list to store the within-cluster sum of squares
wcss = []

# Run K-means for each cluster number and calculate WCSS
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    for k in k_values:
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(X)
        wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(k_values, wcss, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method for Optimal k')
plt.show()
```

From the graph above, and the fact the we have 4 family of frog species, I would choose the number of cluster = 4.

```python
from sklearn.decomposition import PCA
import plotly.express as px

# Perform PCA for dimensionality reduction to 3 dimensions
pca = PCA(n_components=3)
pca_X = pca.fit_transform(X)

colors = np.array([x for x in 'bgrcmykbgr'])

# Run K-means clustering on the transformed data
km_pca = KMeans(n_clusters=4).fit(pca_X)

# Create a 3D scatter plot
fig = px.scatter_3d(x=pca_X[:, 0], y=pca_X[:, 1], z=pca_X[:, 2], color=colors[km_pca.labels_])

fig.show()
```

```python
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score

# Assuming you have performed K-means clustering and obtained the cluster labels
cluster_labels = kmeans.labels_

# Assuming you have the original 'Family' labels in a separate variable
original_labels = df_final['Family']

# Calculate the Adjusted Rand Index
ari = adjusted_rand_score(original_labels, cluster_labels)
print("Adjusted Rand Index:", ari)
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import silhouette_score
import warnings

# Suppress all warnings
warnings.filterwarnings("ignore")

# Extract the MFCC columns from the df_final DataFrame
X = df_final[mfcc_columns]

# Define the parameter grid
param_grid = {
    'eps': [0.5, 1.0, 1.5],
    'min_samples': [3, 5, 7]
}

# Define a custom scoring function for silhouette score
def calculate_silhouette(estimator, X):
    cluster_labels = estimator.fit_predict(X)
    return silhouette_score(X, cluster_labels)

# Perform grid search with cross-validation
dbscan = DBSCAN()
grid_search = GridSearchCV(dbscan, param_grid, scoring=calculate_silhouette)
grid_search.fit(X)

# Get the best parameters and best silhouette score
best_params = grid_search.best_params_
best_score = grid_search.best_score_
print("Best Parameters:", best_params)
print("Best Silhouette Score:", best_score)
```
